```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.collections import PolyCollection
import matplotlib as mpl

mpl.rcParams['figure.figsize'] = (12,12)
```

```python
# set up offsets for hexagon vertices
theta = np.pi/6 + np.arange(6)*np.pi/3
offsets = np.zeros( (6,2) )
offsets[:,0] = np.cos(theta)/np.sqrt(3)
offsets[:,1] = np.sin(theta)/np.sqrt(3)

# rows and columns of hexagons
rows = 50
cols = 50

# number of hexagons
n = rows*cols

# initialize the centers of the hexagons
x = 1.0*np.arange(cols)
y = np.arange(rows)*np.sqrt(3)/2
X,Y = np.meshgrid(x,y)
X[1::2,:] += 0.5

# centers of hexagons
centers = np.vstack( [X.ravel(), Y.ravel()] ).T

# place a hexagon at each center
verts = [ centers[i,:] + offsets for i in range(n) ]
```

```python
# flip a coin for each hexagon
p = 0.5
z = np.random.binomial(1,p,(n,))

# form the hexagon collection
hexes = PolyCollection(verts, linewidths=0, array=z, cmap='Blues')

# plot
plt.gca().add_collection(hexes)
plt.gca().add_collection(hexes)
plt.xlim( [-0.5,cols] )
plt.ylim( [-1./np.sqrt(3),(rows-1)*np.sqrt(3)/2+1./np.sqrt(3)] )

None
```
